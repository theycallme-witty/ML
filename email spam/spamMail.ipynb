```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
spamData= pd.read_csv('./mail_data.csv')
spamData.head()
```

```python
mail_data= spamData.where((pd.notnull(spamData)),'') 
```

```python
mail_data.shape
```

```python
mail_data.loc[mail_data['Category']=='spam','Category',]=0
mail_data.loc[mail_data['Category']=='ham','Category',]=1
```

```python
X=mail_data['Message']
Y=mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase='True')

X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)

Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
print(X_train_features)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train_features,Y_train)
```

```python
prediction_training=model.predict(X_train_features)
accuracy_training=accuracy_score(Y_train,prediction_training)
```

```python
print("the accuracy of training data is :",accuracy_training)
```

```python
prediction_test=model.predict(X_test_features)
accuracy_tesst=accuracy_score(Y_test,prediction_test)
```

```python
print("the accuracy of test data is :", accuracy_tesst)
```

```python
input_mail=["I'm gonna be home soon and i don't want to talk about this stuff anymore tonight, k? I've cried enough today."]
tranform_data=feature_extraction.transform(input_mail)
prediction_data=model.predict(tranform_data)
# print(prediction_data)
if prediction_data[0]==1:
    print("it is an ham mail")
else:
    print("it is an spam mail")    
```
