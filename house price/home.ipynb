```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv("./Housing.csv")
data.isnull().sum()
data.drop(['prefarea'],axis='columns',inplace=True)
conv_dates=[1 if values<3 else 0 for values in data.bedrooms]
data['bedrooms']=conv_dates
```

```python
data.head()
```

```python
data.info()
```

```python
data1=pd.get_dummies(data['furnishingstatus'])
data=pd.concat([data,data1],axis=1).reindex(data.index)
data.drop('furnishingstatus',axis=1,inplace=True)
data
```

```python
y=data[['price']].values
x=data.drop(['price'],axis=1).values
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2)
```

```python
lire=LinearRegression()
lire.fit(x_train,y_train)
```

```python
y_predict=lire.predict(x_test)
print(y_predict)
print(y_test)
```

```python
print('coefficient of determination: %.3f'% r2_score(y_test,y_predict))
```
